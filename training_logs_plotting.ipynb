```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
SAC_safe = pd.read_csv('training logs/SAC_safe.csv')
SAC_agg = pd.read_csv('training logs/SAC_aggressive.csv')
SAC_agg_br = pd.read_csv('training logs/SAC_aggressive_withbrakes.csv')

```

```python
# Display the first few rows of the dataframe to understand its structure
print(SAC_safe.head())

# Plotting rewards vs timesteps
plt.figure(figsize=(10, 5))
plt.plot(SAC_safe['time/total_timesteps'], SAC_safe['rollout/ep_rew_mean'], marker='o', linestyle='-', color='b')
plt.title('SAC safe')
plt.xlabel('Total Timesteps')
plt.ylabel('Average Episode Reward')
plt.grid(True)
plt.show()



# print(SAC_agg.head())

# Plotting rewards vs timesteps
plt.figure(figsize=(10, 5))
plt.plot(SAC_agg['time/total_timesteps'], SAC_agg['rollout/ep_rew_mean'], marker='o', linestyle='-', color='b')
plt.title('SAC aggressive')
plt.xlabel('Total Timesteps')
plt.ylabel('Average Episode Reward')
plt.grid(True)
plt.show()



# print(SAC_agg_br.head())

# Plotting rewards vs timesteps
plt.figure(figsize=(10, 5))
plt.plot(SAC_agg_br['time/total_timesteps'], SAC_agg_br['rollout/ep_rew_mean'], marker='o', linestyle='-', color='b')
plt.title('SAC aggressive with brakes')
plt.xlabel('Total Timesteps')
plt.ylabel('Average Episode Reward')
plt.grid(True)
plt.show()
```
